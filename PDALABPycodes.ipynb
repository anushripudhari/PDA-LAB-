```python
import pandas as pd
import numpy as np

# -------- Function 1: Handle Missing Values --------
def handle_missing(area, price):
    print("\n--- Handling Missing Values ---")
    
    area = [np.nan if x == -1 else x for x in area]
    price = [np.nan if x == -1 else x for x in price]
    
    df = pd.DataFrame({"Area": area, "Price": price})
    
    df["Area"] = df["Area"].fillna(df["Area"].mean())
    df["Price"] = df["Price"].fillna(df["Price"].mean())
    
    print(df)

# -------- Function 2: Noise Removal --------
def remove_noise(area, price):
    print("\n--- Removing Noise (Outliers) ---")
    
    df = pd.DataFrame({"Area": area, "Price": price})
    
    df = df[(df["Area"] < 5000) & (df["Price"] < 10000000)]
    
    print(df)

# -------- Function 3: Remove Redundancy --------
def remove_duplicates(area, price):
    print("\n--- Removing Duplicate Data ---")
    
    df = pd.DataFrame({"Area": area, "Price": price})
    
    df = df.drop_duplicates()
    
    print(df)

# -------- Main Program --------
n = int(input("Enter number of houses: "))

print("Enter area values:")
area = list(map(float, input().split()))

print("Enter price values:")
price = list(map(float, input().split()))

if len(area) != len(price):
    print("Error: Length mismatch")
    exit()

print("\nChoose Operation:")
print("1. Handle Missing Values")
print("2. Remove Noise")
print("3. Remove Duplicates")

choice = int(input("Enter choice (1/2/3): "))

if choice == 1:
    handle_missing(area, price)
elif choice == 2:
    remove_noise(area, price)
elif choice == 3:
    remove_duplicates(area, price)
else:
    print("Invalid choice")
```

```python
import pandas as pd

# -------- Function: Correlation --------
def find_correlation(area, price):
    print("\n--- Calculating Correlation ---")
    
    df = pd.DataFrame({"Area": area, "Price": price})
    
    corr = df["Area"].corr(df["Price"])
    
    print("Correlation value:", corr)

# -------- Main Program --------
n = int(input("Enter number of houses: "))

print("Enter area values:")
area = list(map(float, input().split()))

print("Enter price values:")
price = list(map(float, input().split()))

# Check length
if len(area) != len(price):
    print("Error: Length mismatch")
    exit()

find_correlation(area, price)

    
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# -------- Function --------
def linear_regression(area, price):
    print("\n--- Linear Regression (Using Library) ---")
    
    # Convert to numpy arrays
    X = np.array(area).reshape(-1, 1)
    y = np.array(price)
    
    # Create model
    model = LinearRegression()
    
    # Train model
    model.fit(X, y)
    
    # Get slope and intercept
    m = model.coef_[0]
    b = model.intercept_
    
    print("Slope (m):", m)
    print("Intercept (b):", b)
    
    # Prediction
    val = float(input("Enter area to predict price: "))
    pred = model.predict([[val]])
    
    print("Predicted Price:", pred[0])

# -------- Main --------
n = int(input("Enter number of houses: "))

print("Enter area values:")
area = list(map(float, input().split()))

print("Enter price values:")
price = list(map(float, input().split()))

if len(area) != len(price):
    print("Error: Length mismatch")
    exit()


linear_regression(area, price)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# -------- Function --------
def logistic_regression(hours, result):
    print("\n--- Logistic Regression ---")
    
    # Convert to numpy arrays
    X = np.array(hours).reshape(-1, 1)
    y = np.array(result)
    
    # Create model
    model = LogisticRegression()
    
    # Train model
    model.fit(X, y)
    
    # Take input for prediction
    val = float(input("Enter study hours to predict Pass/Fail: "))
    
    pred = model.predict([[val]])
    prob = model.predict_proba([[val]])
    
    print("Prediction (0=Fail, 1=Pass):", pred[0])
    print("Probability:", prob[0])

# -------- Main --------
n = int(input("Enter number of students: "))

print("Enter study hours:")
hours = list(map(float, input().split()))

print("Enter results (0=Fail, 1=Pass):")
result = list(map(int, input().split()))

if len(hours) != len(result):
    print("Error: Length mismatch")
    exit()

logistic_regression(hours, result)
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

# -------- Function --------
def decision_tree(temp, condition):
    print("\n--- Decision Tree (Cold vs Fever) ---")
    
    X = np.array(temp).reshape(-1, 1)
    y = np.array(condition)
    
    model = DecisionTreeClassifier()
    model.fit(X, y)
    
    # Prediction
    val = float(input("Enter temperature to predict condition: "))
    pred = model.predict([[val]])
    
    if pred[0] == 0:
        print("Prediction: Cold")
    else:
        print("Prediction: Fever")

# -------- Main --------
n = int(input("Enter number of patients: "))

print("Enter temperatures:")
temp = list(map(float, input().split()))

print("Enter condition (0=Cold, 1=Fever):")
condition = list(map(int, input().split()))

if len(temp) != len(condition):
    print("Error: Length mismatch")
    exit()

decision_tree(temp, condition)
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# -------- Function --------
def random_forest(temp, condition):
    print("\n--- Random Forest Classification ---")
    
    X = np.array(temp).reshape(-1, 1)
    y = np.array(condition)
    
    # Create model (multiple trees)
    model = RandomForestClassifier(n_estimators=10)
    
    # Train model
    model.fit(X, y)
    
    # Prediction
    val = float(input("Enter temperature to predict condition: "))
    pred = model.predict([[val]])
    
    if pred[0] == 0:
        print("Prediction: Cold")
    else:
        print("Prediction: Fever")

# -------- Main --------
n = int(input("Enter number of patients: "))

print("Enter temperatures:")
temp = list(map(float, input().split()))

print("Enter condition (0=Cold, 1=Fever):")
condition = list(map(int, input().split()))

if len(temp) != len(condition):
    print("Error: Length mismatch")
    exit()

random_forest(temp, condition)
```

```python
# Descriptive Analysis (Healthcare Data)

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# -------- Dataset --------
data = {
    "Age": [25, 30, 35, 40],
    "BP": [120, 130, 125, 140]
}

df = pd.DataFrame(data)

# -------- Display Data --------
print("Healthcare Data:\n", df)

print("\n--- Data Info ---")
df.info()

print("\n--- Shape ---")
print(df.shape)

print("\n--- Descriptive Analysis ---")
desc = df.describe()
print(desc)

# -------- Explain describe --------
print("\n--- Key Statistics ---")
print("Mean Age:", desc.loc["mean", "Age"])
print("Mean BP:", desc.loc["mean", "BP"])
print("Max BP:", desc.loc["max", "BP"])
print("Min BP:", desc.loc["min", "BP"])

# -------- Missing Values --------
print("\n--- Missing Values ---")
print(df.isnull().sum())

# -------- Visualization --------
plt.figure()
plt.plot(df['Age'], df['BP'], marker='o')
plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("BP")
plt.show()

plt.figure()
plt.scatter(df['Age'], df['BP'])
plt.title("Scatter: Age vs BP")
plt.xlabel("Age")
plt.ylabel("BP")
plt.show()

# -------- Insights --------
print("\n--- Insights ---")
print("Average Age:", df["Age"].mean())
print("Max BP:", df["BP"].max())
print("Min BP:", df["BP"].min())
```

```python
# Predictive Analysis (Weather Forecast)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# -------- Dataset --------
data = {
    "Temperature": [20, 22, 25, 28, 30, 35],
    "Rain": [1, 1, 1, 0, 0, 0]
}

df = pd.DataFrame(data)

# -------- Display Data --------
print("Weather Data:\n", df)

print("\n--- Data Info ---")
df.info()

print("\n--- Shape ---")
print(df.shape)

print("\n--- Description ---")
desc = df.describe()
print(desc)

# -------- Visualization --------
plt.figure()
plt.scatter(df["Temperature"], df["Rain"])
plt.title("Temperature vs Rain")
plt.xlabel("Temperature")
plt.ylabel("Rain (0/1)")
plt.show()

# -------- Model Building --------
X = df[["Temperature"]]
y = df["Rain"]

model = LogisticRegression()
model.fit(X, y)

# -------- Prediction --------
temp = float(input("\nEnter temperature to predict rain: "))
pred = model.predict([[temp]])

if pred[0] == 1:
    print("Prediction: Rain")
else:
    print("Prediction: No Rain")
```
